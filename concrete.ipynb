```python
import pandas as pd
import numpy as np
```

```python
cement = pd.read_csv('concrete.csv')
```

```python
table1 = pd.DataFrame(cement)
```

```python
table1
```

```python
table1.head()
```

```python
table1 = table1.dropna()
table1
```

```python
table = pd.DataFrame()
table['cement'] = table1['cement']
table['water'] = table1['water']
table
```

```python
mean_of_cement = table['cement'].mean()
mean_of_water = table['water'].mean()
print(mean_of_cement)
print(mean_of_water)
```

```python
table['cement-mean_of_cement'] = table['cement']-mean_of_cement
```

```python
print(table)
```

```python
table['water-mean_of_water'] = table['water'] - mean_of_water
```

```python
print(table)
```

```python
table
```

```python
del table['product of cement-mean_of_cement and water-mean_of_water']
```

```python
table
```

```python
table['product of cement-mean_of_cement and water-mean_of_water'] = table['cement-mean_of_cement'] * table['water-mean_of_water']
```

```python
table
```

```python
sumo = table['product of cement-mean_of_cement and water-mean_of_water'].sum()
print(sumo)
```

```python
sumo1 = table['cement-mean_of_cement'].sum()
print(sumo1)
```

```python
m = sumo/sumo1
print(m)
```

```python
y_intercept = mean_of_water - m*mean_of_cement
```

```python
print(y_intercept)
```

```python
table['y_predicted'] = m*table['cement'] + y_intercept
```

```python
table
```

```python
table['water - mean_of_water'] = table['water'] - mean_of_water
```

```python
table['square of water-mean_of_water'] = table['water - mean_of_water'] * table['water - mean_of_water']
```

```python
table
```

```python
table['y_predicted - mean_of_water'] = table['y_predicted']-mean_of_water
```

```python
table
```

```python
table['square of y_predicted - mean_of_water'] = table['y_predicted - mean_of_water'] * table['y_predicted - mean_of_water']
```

```python
table
```

```python
sum1 = table['square of y_predicted - mean_of_water'].sum()
print(sum1)
```

```python
table['square of water-mean_of_water'] = table['water - mean_of_water'] * table['water - mean_of_water']
table
```

```python
sum1 = table['square of y_predicted - mean_of_water'].sum()
```

```python
sum2 = table['square of water-mean_of_water'].sum()
```

```python
r_square = sum1/sum2
print(r_square)
```
