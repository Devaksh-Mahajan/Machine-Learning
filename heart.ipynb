```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as lib
import seaborn as sns
```

```python
table = pd.read_csv("HeartDisease.csv")
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
 from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard = StandardScaler()
```

```python
column = ['age', 'rest_bps', 'cholestrol', 'thalach', 'old_peak']
```

```python
table[column] = standard.fit_transform(table[column])
```

```python
table.head()
```

```python
y = table['target']
```

```python
x = table.drop(['target'], axis = 1)
```

```python
clf = clf.fit(x, y)
```

```python
tree.plot_tree(clf)
```

```python
clf.feature_importances_
```

```python
clf.get_depth()
```

```python
clf.get_n_leaves()
```

```python
clf.score(x, y, sample_weight = None) #also returns the accuracy
```

```python
pred = clf.predict(x)
```

```python
pred
```

```python
lib.plot(pred, table['age'])
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y, pred))
```
